This is a simple piece of code. It takes in root files that have had the coinc module run with them. It then displays and fits the required plots for time of flight. The information you need is the centriods from the twc corrrected plots. Use those for tof calculations.

```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

# === USER SETTINGS ===
rootPath   = "/Users/sidne/DL_rootfiles/"
runNumber  = "00787"   # change if needed
fileName   = f"{rootPath}output{runNumber}.root"

target_dir = "coinc16"
hist_prefix = "coinc_tof_cpair"
plot_each = True   # set False to skip plotting
save_csv  = True   # set False if you do not want a CSV

# === MODEL ===
def gauss_with_offset(x, A, mu, sigma, C):
    # y = A * exp(-(x - mu)^2 / (2*sigma^2)) + C
    return A * np.exp(-0.5 * ((x - mu) / np.maximum(sigma, 1e-12))**2) + C

results = []

with uproot.open(fileName) as f:
    # iterate objects in the directory, use items() to avoid cycle suffix issues
    try:
        dir_obj = f[target_dir]
    except KeyError:
        raise RuntimeError(f"Directory '{target_dir}' not found in {fileName}")

    for name, obj in dir_obj.items():
        if not name.startswith(hist_prefix):
            continue
        if not hasattr(obj, "to_numpy"):
            continue  # skip non-TH1 objects

        # get histogram as numpy
        y, x_edges = obj.to_numpy()
        if y is None or x_edges is None:
            continue

        # only look at histograms that contain counts
        total_counts = np.sum(y)
        if total_counts <= 0:
            print(f"Skipping {name}: empty histogram.")
            continue

        # prepare x as bin centers
        x = 0.5 * (x_edges[:-1] + x_edges[1:])
        # use only bins with counts for robust initial guesses
        nz = y > 0
        if np.count_nonzero(nz) < 3:
            print(f"Skipping {name}: too few nonzero bins for a stable fit.")
            continue

        x_nz = x[nz]
        y_nz = y[nz]

        # initial guesses
        A0 = float(np.max(y_nz) - np.median(y_nz))
        A0 = max(A0, 1.0)
        mu0 = float(x_nz[np.argmax(y_nz)])
        # weighted std as sigma guess, fall back if needed
        try:
            w = y_nz / np.sum(y_nz)
            sigma0 = float(np.sqrt(np.sum(w * (x_nz - np.sum(w * x_nz))**2)))
            if not np.isfinite(sigma0) or sigma0 <= 0:
                sigma0 = (x_edges[-1] - x_edges[0]) / 10.0
        except Exception:
            sigma0 = (x_edges[-1] - x_edges[0]) / 10.0

        C0 = float(np.median(y_nz))

        p0 = [A0, mu0, sigma0, C0]

        # bounds to keep fit sensible
        lower = [0.0,  x_edges[0],  (x_edges[1]-x_edges[0]) / 4.0, 0.0]
        upper = [np.inf, x_edges[-1], (x_edges[-1]-x_edges[0]), np.inf]

        # Use sqrt(y) as Poisson-like sigma, avoid zeros
        sigma_y = np.sqrt(np.clip(y, 1.0, None))

        try:
            popt, pcov = curve_fit(
                gauss_with_offset, x, y,
                p0=p0, bounds=(lower, upper),
                sigma=sigma_y, absolute_sigma=True, maxfev=20000
            )
            perr = np.sqrt(np.diag(pcov))
        except Exception as e:
            print(f"Fit failed for {name}: {e}")
            continue

        A, mu, sigma, C = popt
        A_err, mu_err, sigma_err, C_err = perr

        # compute area under the Gaussian component
        area = A * np.sqrt(2.0 * np.pi) * sigma
        # quick GOF: chi2/ndf
        y_fit = gauss_with_offset(x, *popt)
        # avoid dividing by zero in variance
        var = np.clip(sigma_y**2, 1.0, None)
        chi2 = float(np.sum((y - y_fit)**2 / var))
        ndf = max(len(x) - len(popt), 1)
        chi2_ndf = chi2 / ndf

        results.append({
            "hist": name,
            "A": A, "A_err": A_err,
            "mu": mu, "mu_err": mu_err,
            "sigma": sigma, "sigma_err": sigma_err,
            "C": C, "C_err": C_err,
            "area": area,
            "chi2_ndf": chi2_ndf,
            "counts": int(total_counts)
        })

        if plot_each:
            # dense grid for smooth curve
            xx = np.linspace(x_edges[0], x_edges[-1], 2000)
            yy = gauss_with_offset(xx, *popt)

            plt.figure()
            plt.bar(x, y, width=np.diff(x_edges), align="center", edgecolor="black", alpha=0.6)
            plt.plot(xx, yy, lw=2)
            plt.title(f"{name}\nμ={mu:.3f}±{mu_err:.3f}, σ={sigma:.3f}±{sigma_err:.3f}, χ²/ndf={chi2_ndf:.2f}")
            plt.xlabel("x")
            plt.ylabel("Counts")
            plt.xlim(-5, 5)
            plt.tight_layout()
            plt.show()

# Summarize results
df = pd.DataFrame(results).sort_values(by="mu").reset_index(drop=True)
print(df.to_string(index=False))

if save_csv and len(df) > 0:
    out_csv = f"gauss_fits_{target_dir}_{hist_prefix}_{runNumber}.csv"
    df.to_csv(out_csv, index=False)
    print(f"\nSaved: {out_csv}")
```
