```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
# Import the CSV Data
data = pd.read_csv('/Users/gabbygelinas/Desktop/Masters/Convolved_fitting/t14ns_1458')
timeDiff = data['timeDiff']
x_root = data['x_root']

# Define the Gaussian and Square Wave functions with amplitude parameter
def gaussian(x, mu, sigma, A):
    return A * np.exp(-0.5 * ((x - mu) / sigma)**2)

def square_wave(x, width):
    return np.where(np.abs(x) <= width / 2, 1.0, 0.0)

# Define the convolution model with amplitude parameter
def conv_model(params, x):
    gauss = gaussian(x, params[0], params[1], params[3])
    square = square_wave(x, params[2])
    return np.convolve(gauss, square, mode='same')

# Estimate initial guesses for the parameters
max_idx = np.argmax(timeDiff)
mu_initial = x_root[max_idx]  #gaussian mean around the peak
sigma_initial = np.std(x_root)/20  # Standard deviation based on the data spread
width_initial = np.ptp(x_root) / 40  # Width based on a fraction of the data range
amplitude_initial = np.max(timeDiff) / 13  # Initial guess for the amplitude

initial_guess = [mu_initial, sigma_initial, width_initial, amplitude_initial]  # Initial guess array

# Compute the convolution using the initial guess
interp_func = interp1d(x_root, conv_model(initial_guess, x_root), kind='linear', fill_value='extrapolate')
initial_model = interp_func(x_root)

# Plot the original data and the initial guess model
plt.figure()
plt.plot(x_root, timeDiff, 'b', label='Original Data')
plt.plot(x_root, initial_model, 'g', label='Initial Guess Model')
plt.legend()
plt.xlabel('x_root')
plt.ylabel('timeDiff')
plt.title('Original Data and Initial Guess Model')
plt.grid(True)
plt.xlim([np.min(x_root), np.max(x_root)])
plt.show()

# Define the objective function for fitting with amplitude parameter
def objective(params):
    interp_func = interp1d(x_root, conv_model(params, x_root), kind='linear', fill_value='extrapolate')
    model = interp_func(x_root)
    return np.sum((model - timeDiff) ** 2)

# Perform the fit using minimize (similar to fminsearch in MATLAB)
result = minimize(objective, initial_guess)
optimal_params = result.x

# Generate the fitted model
interp_func = interp1d(x_root, conv_model(optimal_params, x_root), kind='linear', fill_value='extrapolate')
fitted_model = interp_func(x_root)

# Display the optimal parameters
print('Optimal Parameters:')
print(f'Mu (mean): {optimal_params[0]}')
print(f'Sigma (standard deviation): {optimal_params[1]}')
print(f'Width: {optimal_params[2]}')
print(f'Amplitude: {optimal_params[3]}')

# Calculate the Full Width at Half Maximum (FWHM)
sigma_fitted = optimal_params[1]
FWHM = 2 * np.sqrt(2 * np.log(2)) * sigma_fitted
print(f'Full Width at Half Maximum (FWHM): {FWHM}')

# Calculate the Residual Sum of Squares (RSS)
RSS = np.sum((timeDiff - fitted_model) ** 2)

# Calculate the Total Sum of Squares (TSS)
TSS = np.sum((timeDiff - np.mean(timeDiff)) ** 2)

# Calculate the Coefficient of Determination (R^2)
R_squared = 1 - (RSS / TSS)

# Calculate the Root Mean Square Error (RMSE)
RMSE = np.sqrt(np.mean((timeDiff - fitted_model) ** 2))

# Display the fit quality metrics
print('Fit Quality Metrics:')
print(f'Residual Sum of Squares (RSS): {RSS}')
print(f'Coefficient of Determination (R^2): {R_squared}')
print(f'Root Mean Square Error (RMSE): {RMSE}')

# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Plot the original data and the fitted model in the first subplot
axs[0].plot(x_root, timeDiff, 'b.', label='Original Data')  # Scatter plot
axs[0].plot(x_root, fitted_model, 'r', label='Fitted Model')
axs[0].legend()
axs[0].set_xlabel('x_root')
axs[0].set_ylabel('timeDiff')
axs[0].set_title('Original Data and Fitted Model (t14ns_1458)')
axs[0].grid(True)
axs[0].set_xlim([-2, 2])

# Calculate the square root of timeDiff for error bars (Poisson error)
error_bars = np.sqrt(timeDiff)

# Plot error bars
axs[0].errorbar(x_root, timeDiff, yerr=error_bars, fmt='b.', capsize=10)

# Calculate the residuals
residuals = (timeDiff - fitted_model) / fitted_model

# Plot the residuals vs fitted values in the second subplot
axs[1].plot(x_root, residuals, 'bo')
axs[1].set_xlabel('x_root')
axs[1].set_ylabel('Normalized Residuals')
axs[1].set_title('Normalized Residuals vs Fitted Values (t14ns_1458)')
axs[1].grid(True)
axs[1].set_xlim([-2, 2])
axs[1].set_ylim([-1, 1])
axs[1].axhline(0, color='red', linestyle='--')  # Add a reference line at y = 0

plt.tight_layout()
plt.show()

# Set the parameters for the Gaussian and Square Wave
mu, sigma, width, A = optimal_params
```
